```python
import numpy as np
import pandas as pd

from sklearn import preprocessing

from keras import models
from keras import layers


TrainDataFile = "C:\\Data\\Titanic\\train.csv"
TestDataFile = "C:\\Data\\Titanic\\test.csv"


TrainData = pd.read_csv(TrainDataFile)
TestData = pd.read_csv(TestDataFile)

TrainData.head()
```

```python
TrainData.info()
```

```python
TrainData.describe()
```

```python
train_y = TrainData.as_matrix(columns=['Survived']).astype(float)
```

```python
train_x = TrainData.drop(['Ticket', 'PassengerId','Survived','Name','Cabin'], axis=1)

#fill missing data
train_x['Embarked'].fillna('S', inplace=True)
train_x['Age'].fillna(train_x['Age'].mean(), inplace=True)

#Handling Categorical data
le = preprocessing.LabelEncoder()
train_x.Sex = le.fit_transform(train_x.Sex)
train_x.Embarked = le.fit_transform(train_x.Embarked)

#Normalize
train_x=(train_x-train_x.min())/(train_x.max()-train_x.min())

train_x.head(10)
```

```python
train_x = train_x.values
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.regularizers import l2
#from keras.utils import np_utils

model = Sequential()
model.add(Dense(256 ,input_dim=7, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu', kernel_regularizer=l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python

epochs = 200
batch_size = 64

#model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
#model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(train_x, train_y, epochs=epochs, batch_size=batch_size, validation_split=0.20)
```
